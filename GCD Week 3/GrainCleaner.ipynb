# 2	Data Validation – choose what values are allowed in a cell  
In this case we don't want any production entries wich contain the word "and", instead we want those products added to the list by removing 'and' and adding a comma. Because we will have to search and replace later the replacement part will be added later on. We will print all the invalid rows

```python
import pandas as pd

data = pd.read_csv('../GCD repo/GCD Week 3/GRAIN_CSV.csv')
production_data = data['Production']
words_to_filter = ['and']
bad_entries = []


rownumber = 0
for row in production_data:
    rowtext = str(row)

    if "and" in rowtext:
        bad_entries.append(rownumber)

    rownumber += 1

data.iloc[bad_entries]['Production']
```

# 4	Find and Replace – look for one thing, and replace it with another
We replace the bad values we found before

```python
import pandas as pd

data = pd.read_csv('../GCD repo/GCD Week 3/GRAIN_CSV.csv')
production_data = data['Production']
words_to_filter = ['and']
bad_entries = []

rownumber = 0
for row in production_data:
    rowtext = str(row)

    if "and" in rowtext:
        rowtext = rowtext.replace('and', ',')
        data.set_value(rownumber, 'Production', rowtext)
        bad_entries.append(rownumber)

    rownumber += 1

data.iloc[bad_entries]['Production']
```

# 5	Paste Special – remove unwanted formatting
We can remove any unwated formatting using simple regex substitions most of the time

```python
import pandas as pd
import re

data = pd.read_csv('../GCD repo/GCD Week 3/GRAIN_CSV.csv')
data['Production']
words = []
# removing all spaces and numbers
for word in data['Production']:
    words.append(re.sub('[^A-Za-z0-9]+', '', str(word)))

words[:10] # Getting only the first 10 to avoid a long list in the notebook
```
